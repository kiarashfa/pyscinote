# Module 02: Plot Types

So far we have worked exclusively with line plots. That is a reasonable starting point because line plots are the workhorse of scientific figures, but Matplotlib supports a wide range of chart types and the right choice depends on what your data is and what question you are asking of it. This module works through the most commonly used types, keeping the underlying data consistent wherever possible so the comparison stays focused on the chart itself rather than on what the numbers mean.

## Imports and Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Shared dataset

Most of the examples in this module use the same underlying data: a smooth curve representing a hypothetical spectral intensity as a function of wavelength. Having one consistent dataset makes it easier to see what each chart type emphasises and what it obscures.

```python
wavelength = np.linspace(400, 700, 30)  # 30 wavelength values from 400 to 700 nm

# A broad Gaussian peak centred at 550 nm, resembling a simple absorption band
intensity = np.exp(-((wavelength - 550) ** 2) / (2 * 60**2))
```

## Line plot

We covered line plots in Module 01. The key characteristic is that the points are connected, implying a continuous relationship between x and y. This is appropriate when your x-axis represents something that varies continuously, like time, temperature, or wavelength.

```python
fig, ax = plt.subplots()

ax.plot(wavelength, intensity)

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Intensity (a.u.)')
ax.set_title('Line plot')
plt.show()
```

The connecting line suggests that something meaningful happens between each measurement. For a smooth physical property like spectral intensity, that is a reasonable assumption.

## Scatter plot

A scatter plot shows the same points with no connecting line. Use `ax.scatter()` instead of `ax.plot()`. The absence of a line makes a different statement: each point stands alone, and you are not implying a smooth function between them. This is the appropriate choice when you are showing individual measurements whose order carries no meaning, or when you want to look for correlations between two variables.

```python
fig, ax = plt.subplots()

ax.scatter(wavelength, intensity, s=40)  # s controls marker size

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Intensity (a.u.)')
ax.set_title('Scatter plot')
plt.show()
```

Same data, different implication. Without the line, the reader sees discrete measurements rather than a continuous curve. Whether that is honest or misleading depends entirely on what the data actually represents.

## Line plot with markers

The two previous styles can be combined: a line with markers at each data point. This is often the most informative choice for experimental data where you have measured values at specific points but the underlying trend is continuous. You get both the individual measurement locations and the overall shape.

```python
fig, ax = plt.subplots()

ax.plot(wavelength, intensity, marker='o', markersize=5)  # Circle at each data point

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Intensity (a.u.)')
ax.set_title('Line plot with markers')
plt.show()
```

The markers show where the actual measurements are; the line shows the trend. Together they let a reader see both the data density and the shape of the relationship.

## Bar chart

Bar charts are for comparing discrete categories or a small number of distinct values. They do not imply continuity between bars the way a line does. We will use a subset of our wavelength data here to keep the chart readable: ten wavelength bins and their corresponding intensities, treated as if they were discrete sample points.

```python
# Take every third point to get a sparser set suitable for a bar chart
wl_subset = wavelength[::3]
int_subset = intensity[::3]

fig, ax = plt.subplots()

ax.bar(wl_subset, int_subset, width=18)  # Controls width size
# ax.bar(wl_subset, int_subset, width=18, color='steelblue', edgecolor='white')  # More options

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Intensity (a.u.)')
ax.set_title('Vertical bar chart')
plt.show()
```

The `width` argument controls how wide each bar is. Setting it close to the spacing between x values fills the chart; setting it smaller leaves gaps. The `edgecolor='white'` adds a thin white border between bars, which helps when bars are adjacent.

A horizontal bar chart uses `ax.barh()` instead of `ax.bar()`, and the x and y roles swap. This is often more readable when category labels are long.

```python
fig, ax = plt.subplots()

# barh takes (y_positions, bar_lengths)
ax.barh(wl_subset, int_subset, height=18, color='steelblue', edgecolor='white')

ax.set_ylabel('Wavelength (nm)')
ax.set_xlabel('Intensity (a.u.)')
ax.set_title('Horizontal bar chart')
plt.show()
```

The data is identical in both bar charts; only the orientation changes. In this case neither is obviously better, but with longer category labels the horizontal version is usually cleaner because the labels have room to sit beside the bars rather than below them.

## Histogram

A histogram shows the distribution of values in a dataset. It answers the question: how often does each value (or range of values) occur? This makes it the natural choice when you have many repeated measurements and you want to understand the spread and shape of the data.

To demonstrate this properly, we need data that actually has a distribution. We will generate a synthetic dataset representing repeated measurements of a physical quantity, such as the absorption peak position from a set of replicate experiments. Real repeated measurements scatter around a true value with some noise, and that scatter follows a roughly normal distribution.

```python
# Setting a seed makes this notebook reproducible: anyone who runs it gets the same random numbers.
# Without a seed, the numbers change every time the cell runs, which makes it hard to share and debug.
np.random.seed(42)

# 200 simulated measurements of a peak position, centred at 550 nm
# with a standard deviation of 8 nm representing typical instrument noise
peak_measurements = np.random.normal(loc=550, scale=8, size=200)
```

`np.random.normal(loc, scale, size)` draws values from a normal distribution with mean `loc` and standard deviation `scale`. With `size=200` we get 200 independent draws, which is a reasonable approximation of what a series of replicate measurements looks like when instrument noise is the dominant source of variation.

```python
fig, ax = plt.subplots()

ax.hist(peak_measurements, bins=25, color='steelblue', edgecolor='white')

ax.set_xlabel('Peak position (nm)')
ax.set_ylabel('Count')
ax.set_title('Distribution of replicate peak measurements')
plt.show()
```

The bell shape is visible but not perfectly smooth because we have 200 points, not 20000. The `bins` argument controls how many intervals the data range is divided into: too few and you lose the shape; too many and each bar has so few counts that the histogram looks noisy. For 200 data points, somewhere between 15 and 30 bins usually works well.

## Pie chart

Pie charts need categorical data that sums to a meaningful whole. The wavelength dataset does not qualify, so we switch to a small industry dataset: the approximate global production share of the five most common thermoplastics.

```python
# Approximate global production share of the five most common thermoplastics
components     = ['PE', 'PP', 'PVC', 'PET', 'PS']
weight_percent = [32.0, 24.0, 19.0, 14.0, 11.0]  # values sum to 100

# The bare essentials way
fig, ax = plt.subplots()
ax.pie(weight_percent, labels=components)
plt.show()
```

```python
# Nicer way
fig, ax = plt.subplots()

wedges, texts, autotexts = ax.pie(
    weight_percent,     # same as before
    labels=components,  # same as before
    autopct='%1.1f%%',  # percentage inside each slice
    startangle=90,      # Rotates the start to the top (12 o'clock)
    pctdistance=0.6     # keep percentage text well inside the wedge
)

# even nicer (you can remove labels and add them as legend by uncommenting the line below)
# ax.legend(wedges, components, title='Polymer', loc='lower right', bbox_to_anchor=(1.15, 0.0), frameon=False)

ax.set_title('Global polymer production by share (approximate)')
plt.show()
```

Moving the category names into a legend and keeping only the percentages inside the wedges is a practical way to avoid overlap, especially when several slices are similarly sized. The `pctdistance` argument controls how far the percentage label sits from the center: values below 1.0 place it inside the slice. Note that we have not yet touched font sizes, colors, or label styling; those tools are all covered in Module 03.

## Radar Chart (Spider Plot)

A radar chart places each measured property on its own axis radiating from the center, and connects the values into a polygon. The shape of that polygon is the signature of the material: a sample that scores high on everything produces a large polygon, while a narrow polygon in one direction reveals a weakness. This makes radar charts useful for comparing two materials side by side, particularly when each has different strengths, which is a common situation when benchmarking a new formulation against a commercial reference.

```python
# Mechanical properties for PP and HDPE, normalized to a 0-100 scale (100 = best in class)
# Raw reference values:
#   Tensile Strength (MPa):    PP ~35,  HDPE ~28
#   Elongation at Break (%):   PP ~150, HDPE ~700
#   Flexural Modulus (GPa):    PP ~1.5, HDPE ~1.0
#   Hardness (Shore D):        PP ~75,  HDPE ~68
#   Impact Strength (kJ/m2):   PP ~5,   HDPE ~15
#   Heat Resistance / HDT (C): PP ~100, HDPE ~75

categories = [
    'Tensile\nStrength',
    'Elongation\nat Break',
    'Flexural\nModulus',
    'Hardness\n(Shore D)',
    'Impact\nStrength',
    'Heat\nResistance'
]

pp_values   = [85, 35, 90, 88, 40, 95]   # stiff, strong, heat-resistant, brittle
hdpe_values = [68, 95, 62, 80, 85, 72]   # tough, flexible, impact-resistant

# 1. Calculate angles and "close the loop"
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]
pp_values += pp_values[:1]
hdpe_values += hdpe_values[:1]

# 2. Create the plot
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'))

# Plot and fill
ax.plot(angles, pp_values, label='PP')
ax.fill(angles, pp_values, alpha=0.2)
ax.plot(angles, hdpe_values, label='HDPE')
ax.fill(angles, hdpe_values, alpha=0.2)

# 3. Add category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1.10))
plt.show()
```

PP scores higher on stiffness, modulus, and heat resistance, while HDPE scores higher on toughness and elongation at break, which matches their known behavior in real applications. The scores are normalized relative to each other on a 0 to 100 scale, not raw physical units; the raw reference values are documented in the code comments. Font sizes, colors, and label styling are all adjustable and will be covered properly in Module 03.

Radar charts work best with 5 to 8 properties; beyond that the chart becomes difficult to read. Also keep in mind that the enclosed area scales with the square of the values, so a material that scores uniformly at 80 across all axes will look disproportionately larger than one that scores 40, even though the difference in each individual property is the same.

## Choosing the right chart type

The choice of chart type is not just aesthetic. It shapes what a reader can and cannot extract from your figure, so it is worth thinking about deliberately.

Line plots are appropriate when x is continuous and the relationship between x and y is the point. They imply that something is happening between your measured points, so if that implication is false, a line is misleading. Scatter plots make no such claim: each point is an observation, and the reader is left to judge whether a pattern exists. Line plots with markers give you both, which is often the honest choice for experimental data where you have discrete measurements of a continuous process.

Bar charts are for comparisons across discrete categories or conditions. They work well when you have a small number of groups and want to compare magnitudes directly. They are misused when applied to continuous x-axis data, where a line would be more appropriate.

Histograms are for distributions. They answer a different question than the other chart types: not "what is the value at each point" but "how are the values spread across the range." Whenever you have a large set of repeated measurements and you want to characterise the variability, a histogram is the first thing to reach for.

Pie charts deserve some caution. They are genuinely useful for one narrow purpose: showing the part-to-whole relationship for a small number of categories when the proportions are what matter. The problem is that human vision is not good at comparing angles and arc lengths, so if you need your reader to compare the sizes of slices accurately, a bar chart will almost always serve them better. Pie charts also break down entirely with more than about five categories. Reserve them for cases where the message is simple enough that approximate proportions are sufficient.
