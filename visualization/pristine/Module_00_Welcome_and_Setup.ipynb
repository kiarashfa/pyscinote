## How to use this course

Each module is a self-contained notebook. Within a module, cells are meant to be run in order, from top to bottom. Later cells often depend on variables or imports from earlier ones, so skipping ahead will produce errors. If something goes wrong mid-notebook, the most reliable fix is to go to **Runtime > Restart and run all** and let it rebuild from scratch.

To run a cell, click on it and press **Shift+Enter**. The cell executes, any output appears directly below it, and the cursor moves to the next cell. You can also use the play button on the left side of each cell, but Shift+Enter is faster once you are used to it.

## What Python is, and why scientists use it

Python is a general-purpose programming language that has become the default tool for data analysis and scientific computing. It did not win that position because it is the fastest language or the most powerful, but because it has a large, well-maintained ecosystem of libraries for exactly the kinds of things scientists need: numerical computation, data manipulation, plotting, statistics, and machine learning. The code tends to read clearly, which makes it easier to share, check, and pick back up after six months away.

Python is superior because it provides a completely free, open-source environment where you can build consistent, code-based visualizations that integrate seamlessly with any data source or workflow without the cost and limitations of proprietary licenses.

## A map of the libraries in this course

You do not need to know any of these libraries yet. This is just an orientation: what each one does and roughly when it appears. They will all make sense in context when the time comes.

- **NumPy**: The foundation for numerical work in Python; it gives you fast arrays and a wide range of mathematical operations.
- **pandas**: Handles tabular data (think spreadsheets) in a way that makes loading, filtering, and transforming datasets concise and readable.
- **Matplotlib**: The core plotting library; almost every scientific figure you have seen produced in Python was made with Matplotlib or something built on top of it.
- **SciPy**: Extends NumPy with scientific algorithms, including the curve-fitting tools we use in Module 04.
- **Seaborn**: A statistical visualization library built on Matplotlib; it makes distribution plots, heatmaps, and multi-variable summaries much less work.
- **Plotly**: Produces interactive, browser-based figures where you can zoom, hover, and explore data; introduced in Module 10.
- **RDKit**: A cheminformatics library used to draw 2D molecular structures; introduced in Module 12.
- **py3Dmol**: Renders interactive 3D molecular structures directly inside a notebook; also introduced in Module 12.

## The Google Colab interface

If you are opening this notebook in Colab for the first time, here is what you need to know about the environment.

**Running cells:** Click any cell and press `Shift+Enter`. The brackets on the left of a code cell change from `[ ]` to `[*]` while the cell is running, and then show a number (e.g., `[1]`) once it has finished. That number is the execution order, which can help you track what has run.

**Adding and removing cells:** To add a new cell, hover your mouse over the space between existing cells and click on **+ Code** or **+ Text**. You can also use `Ctrl + M` followed by `A` (to add above) or `B` (to add below). To remove a block, click the **trash can icon** on the cell toolbar or press `Ctrl + M` then `D`.

**Running cells out of order:** Colab lets you run any cell at any time, but within these notebooks you should not. Variables and imports accumulate in memory in the order you run cells. If you run a cell that uses a variable before running the cell that defines it, Python will raise a `NameError`. When in doubt, restart.

**Restarting the runtime:** Go to **Runtime > Restart session** in the menu bar. This clears all variables from memory and resets the Python environment. Your notebook file is not affected. After restarting, run all cells again from the top. The shortcut **Runtime > Restart session and run all** does this in one step.

**The Files panel:** On the left sidebar, click the folder icon to open the Files panel. This shows the file system of your Colab session. You can upload files by clicking the upload button (the page icon with an upward arrow). Files you upload here are available to your notebook as if they were on your local machine, at the path `/content/your_filename.csv` or wherever they land. Note that this file system is temporary: when your session ends, uploaded files are gone unless you save them to Google Drive.

## Setting up the data folder

Several modules in this course load data from CSV or Excel files. The convention throughout the course is that these files live in a folder called `data/` inside your Colab session. Creating that folder takes two seconds.

Open the Files panel on the left sidebar (the folder icon). You will see a folder called `sample_data` that Colab creates by default. To create a new folder, click the folder-with-a-plus icon at the top of the Files panel and name it `data`. That is all. When you upload a data file for a module, upload it into that `data/` folder.

You can also create the folder from a code cell, which is sometimes faster:

````python
import os

os.makedirs('data', exist_ok=True)  # Creates the folder; does nothing if it already exists
print('data/ folder is ready')
````

The `exist_ok=True` argument means this cell is safe to run more than once; it will not raise an error if the folder already exists.

## Smoke test: checking your environment

The cell below imports the core libraries and prints their version numbers. Run it now. If everything is working, you will see a clean list of versions. Colab comes with most of these pre-installed, so errors here are uncommon, but they do occasionally happen after a runtime update.

````python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import plotly

print(f'numpy     {np.__version__}')
print(f'pandas    {pd.__version__}')
print(f'matplotlib {matplotlib.__version__}')
print(f'seaborn   {sns.__version__}')
print(f'plotly    {plotly.__version__}')
````
